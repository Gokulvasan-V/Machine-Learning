```python
from bs4 import BeautifulSoup # this is collecting data from internet
from selenium import webdriver # it is for pulling data out of HTML & HML files

import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
Product=[]
Price=[]
Rating=[]
```

```python
driver.get("https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3DApple&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIlNob3AgTm93Il0sInZhbHVlVHlwZSI6Ik1VTFRJX1ZBTFVFRCJ9fX19fQ%3D%3D&otracker=clp_metro_expandable_8_3.metroExpandable.METRO_EXPANDABLE_Shop%2BNow_mobile-phones-store_92RED14GXPXF_wp2&fm=neo%2Fmerchandising&iid=M_0f3ec54a-3a62-4d01-ba19-2b870b650b96_3.92RED14GXPXF&ppt=hp&ppn=homepage&ssid=2qq2arkzao0000001675218998843")
```

```python
content=driver.page_source # collecting HTML content

content
```

```python
soup=BeautifulSoup(content) # convert structure format of HTML content

soup
```

```python
for i in soup.findAll('a',href=True,attrs={'class':"_1fQZEK"}):
    name=i.find('div',attrs={'class':'_4rR01T'})
    price=i.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
    rate=i.find('div',attrs={'class':'_3LWZlK'})
    Product.append(name.text)
    Price.append(price.text)
    Rating.append(rate.text)
```

```python
print(Product[0:7])
print(len(Product))
print('--*'*10)
print(Price[0:7])
print(len(Price))
print('--*'*10)
print(Rating[0:7])
print(len(Rating))
```
